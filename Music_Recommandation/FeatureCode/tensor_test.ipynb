```python
import librosa
import glob
import numpy as np
import librosa.display
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
music_path = '/home/bob/Music/CloudMusic'
```

```python
music_name_list = glob.glob(music_path+'/'+'*.mp3')  
```

```python
rand_music_path = np.random.choice(music_name_list)
```

```python
y, sr = librosa.load(rand_music_path)
```

```python
melspectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
```

```python
D = np.abs(librosa.stft(y))**2
S = librosa.feature.melspectrogram(S=D)
# Passing through arguments to the Mel filters
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128,fmax=8000)
```

```python
plt.figure(figsize=(20, 2.56), dpi=100)
librosa.display.specshow(librosa.power_to_db(S,
                                            ref=np.max), fmax=8000)
plt.margins(0,0)
#plt.savefig(,bbox_inches='tight', pad_inches=0)
```

```python
fig = plt.gcf()
melspec = librosa.feature.melspectrogram(y, sr, n_fft=1024, hop_length=512, n_mels=128)
# convert to log scale
logmelspec = librosa.power_to_db(melspec)

# plot mel spectrogram
librosa.display.specshow(logmelspec, sr=sr)

fig.set_size_inches(25.6 ,2.56)
plt.gca().xaxis.set_major_locator(plt.NullLocator())
plt.gca().yaxis.set_major_locator(plt.NullLocator())
plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, hspace = 0, wspace = 0)
plt.margins(0,0)
#fig.savefig('/home/bob/Desktop/1.png', format='png', 
#            transparent=True, dpi=100, pad_inches = 0)
#plt.close()
```
